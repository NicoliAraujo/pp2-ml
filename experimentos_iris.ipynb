```python
import pandas as pd
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(iris.data)
```

```python
iris_df['target'] = iris['target']
```

```python
iris_df = iris_df.sample(frac=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse=False)
```

```python
iris_df[['target1', 'target2', 'target3']] = ohe.fit_transform(iris_df['target'].reshape(-1,1))
```

```python
iris_df['target1'] = 0
iris_df['target2'] = 0
iris_df['target3'] = 0
```

```python
iris_df
```

```python
import numpy as np
```

```python
y_pred = np.random.random(size=(150,3))
```

```python
y_pred.shape, iris_df[['target1', 'target2', 'target3']].shape
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(np.round(y_pred), iris_df[['target1', 'target2', 'target3']], average='macro')
```
